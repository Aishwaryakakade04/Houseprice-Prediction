```python
df.drop(['user id'],axis=1,inplace=True)     #remove discrete column
```

```python
# Seperate target column
```

```python
x=df.drop(['churn'],axis=1)
y=df['churn']
```

# Preprocessing


```python
for i in x.columns:
    if x[i].dtypes=='object':
        x[i]=x[i].fillna(x[i].mode()[0])
    else:
         x[i]=x[i].fillna(x[i].median())
```

```python
x.isna().sum()
```

```python
# Seperate features
```

```python
x_cat=x.select_dtypes(include='object')
x_con=x.select_dtypes(exclude='object')
```

```python
x_cat
```

```python
x_con
```

# Feature Selection

```python
# Encoding
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
x_cat=x_cat.apply(le.fit_transform)   #label encoding
x_cat
```

```python
# Scaling
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
x_con=pd.DataFrame(ss.fit_transform(x_con),columns=ss.get_feature_names_out())
x_con
```

# Combine categorical and continuous data

```python
x1=x_con.join(x_cat)
```

```python
x1
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
sfs=SequentialFeatureSelector(lr)
sfs.fit_transform(x1,y)
```

```python
cols=sfs.get_feature_names_out()   #list of columns selected by using feature selector.
cols
```

```python
x2=pd.DataFrame(x,columns=cols)   #select the important feature from main dataset by using list of imp column
```

```python
x2
```

# Preprocessing by usning sklearn Pipeline


```python
cat=[]
con=[]

for i in x2.columns:
    if x2[i].dtypes=='object':
        cat.append(i)
    else:
        con.append(i)
```

```python
cat
```

```python
#import libraries for pipeline

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder, StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
# Numerical Pipeline
num_pipeline = Pipeline(steps=[('imputer',SimpleImputer(strategy='mean')),
                               ('scaler',StandardScaler())])

# Categorical Pipeline
cat_pipeline = Pipeline(steps=[('imputer',SimpleImputer(strategy='most_frequent')),
                               ('ord_enc',OrdinalEncoder())])

# Combine pipelines using Column Transformer
pre = ColumnTransformer([('num',num_pipeline,con),
                         ('cat',cat_pipeline,cat)])
```

```python
x3=pre.fit_transform(x2)
x3
```

```python
cols1=pre.get_feature_names_out()
cols1
```

```python
x3=pd.DataFrame(x3,columns=cols1)    #preprocess data after pipeline
x3
```

# Create Training and Testing set


```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x3,y,test_size=0.2,random_state=21)
```

# Model Evaluation


```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.svm import SVC
```

```python
la=LogisticRegression()   
kn=KNeighborsClassifier()
dt=DecisionTreeClassifier()
rf=RandomForestClassifier()
ab=AdaBoostClassifier()
sv=SVC()
```

```python
list1=[la,kn,dt,rf,ab,sv]       
```

```python
for i in list1:                                   #check all model at once
    i.fit(x_train,y_train)                        #fit all model from list
    y_pred=i.predict(x_test)
    from sklearn.metrics import f1_score
    ts_f1 = f1_score(y_pred,y_test)
    
    
    print('__'*20)
    print('Model:',i)
    print('f1_score:',ts_f1)
```

# Hyperparameter Tuning


```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
import numpy as np
```

```python
grid={
    'n_estimators':range(1,300),
    
    'criterion':['gini','entropy'],
    'max_depth':range(1,50),
    'min_samples_split':range(1,10),
    'min_samples_leaf':range(1,10)
}
```

```python
grid
```

# Final Model

```python
rf=RandomForestClassifier(n_estimators=168,min_samples_split=7,min_samples_leaf=4,max_depth=47,criterion='gini')
```

```python
rf.fit(x_train,y_train)
```

# Evaluation


```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

# Training Data Evaluation


```python
y_pred_train=rf.predict(x_train)

print('__'*30)
acc=accuracy_score(y_pred_train,y_train)
print('Accuracy:',acc)

print('__'*30)

cnf=confusion_matrix(y_pred_train,y_train)
print('Confusion_matrix:\n',cnf)

print('__'*30)

clf=classification_report(y_pred_train,y_train)
print('classification report:\n',clf)

print('__'*30)
```

# Testing Data Evaluation


```python
y_pred=rf.predict(x_test)

print('__'*30)
acc1=accuracy_score(y_pred,y_test)
print('Accuracy:',acc1)

print('__'*30)

cnf1=confusion_matrix(y_pred,y_test)
print('Confusion_matrix:\n',cnf1)

print('__'*30)

clf1=classification_report(y_pred,y_test)
print('classification report:\n',clf1)

print('__'*30)
```
